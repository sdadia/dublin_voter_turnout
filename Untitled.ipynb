### Log of DiffAdd

## DiffAdd v/s GenEl2004

As we can see, the p value is less than 0.05, so we can say the variable is important, but the Res v/s Fit plot shows curvature, this is expected since the original distribution shows curvature. The QQ plot is fine.

```python
model1 = smf.ols('GenEl2004 ~ DiffAdd', data=vdata)
model_fit1 = model1.fit()
print(model_fit1.summary())
LinRegDiagnostic(model_fit1).plot_diagnostics()
plt.show()
```

## DiffAdd + log(DiffAdd) v/s GenEl2004

```python
model2 = smf.ols('GenEl2004 ~ DiffAdd + np.log(DiffAdd)', data=vdata)
model_fit2 = model2.fit()
print(model_fit2.summary())
LinRegDiagnostic(model_fit2).plot_diagnostics()
plt.show()
```

## log_DiffAdd  v/s GenEl2004

```python
model3 = smf.ols('GenEl2004 ~ np.log(DiffAdd)', data=vdata)
model_fit3 = model3.fit()
print(model_fit3.summary())
LinRegDiagnostic(model_fit3).plot_diagnostics()
plt.show()
```

## log(DiffAdd) + log(DiffAdd)^2

```python
model4 = smf.ols('GenEl2004 ~ np.log(DiffAdd)  + np.power(np.log(DiffAdd),2)', data=vdata)
model_fit4 = model4.fit()
print(model_fit4.summary())
LinRegDiagnostic(model_fit4).plot_diagnostics()
plt.show()
```

## Poly(log_DiffAdd, 3)

Here we find that the P-value for the third power is > 0.05, thus we do not need this term.
We can confirm this by doing anova in the next step

```python
model5 = smf.ols('GenEl2004 ~ np.log(DiffAdd)  + np.power(np.log(DiffAdd),2) + np.power(np.log(DiffAdd),3)', data=vdata)
model_fit5 = model5.fit()
print(model_fit5.summary())
LinRegDiagnostic(model_fit5).plot_diagnostics()
plt.show()
```

## Anova 
The following anova check tells us that the simpler model (model with only 2 powers) is sufficient

```python
from statsmodels.stats.api import anova_lm

anova_lm(model_fit4, model_fit5)
```
